```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# 데이터 (예시 데이터 사용)
red = 3.0  # red 평균
white = 2.5  # white 평균
red2 = 0.8  # red 표준편차
white2 = 0.7  # white 표준편차

# x축 데이터
wines = ('평균', '표준편차')

# y축 데이터
data = {
    'red': (red, red2),
    'white': (white, white2)
}

# 3D 그래프 설정
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# 막대 그래프의 간격과 색상 설정
x = np.arange(len(wines))
y = np.array([0, 1])  # red와 white의 인덱스를 y축으로 사용
colors = ['#FF7F7F', '#FFFFFF']  # red와 white 막대의 색상

# 각 와인 종류에 대해 3D 막대 그래프 생성
for i, (attribute, measurement) in enumerate(data.items()):
    xpos = x
    ypos = np.full_like(xpos, y[i])
    zpos = np.zeros_like(xpos)
    
    dx = np.ones_like(xpos) * width
    dy = np.ones_like(xpos) * width
    dz = measurement
    
    ax.bar3d(xpos, ypos, zpos, dx, dy, dz, color=colors[i], label=attribute)

# 축 및 레이블 설정
ax.set_title('PH의 평균과 표준편차')
ax.set_xticks(x)
ax.set_xticklabels(wines)
ax.set_yticks(y)
ax.set_yticklabels(['red', 'white'])
ax.set_zlabel('Value')

# 그리드 및 레전드 추가
ax.grid(True, linestyle='--', linewidth=0.5, color='black')
ax.legend(loc='upper right')

plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import scipy.stats as stats

# 예시 데이터 (실제 데이터로 대체 가능)
# red = pd.read_csv('winequality-red.csv')
# white = pd.read_csv('winequality-white.csv')

red_density = red['pH']
white_density = white['pH']

# Bartlett 검정
bartlett_stat, bartlett_p_value = stats.bartlett(red_density, white_density)

# t-검정 (등분산성을 가정하지 않음)
t_stat, p_value = stats.ttest_ind(red_density, white_density, equal_var=False)

# 3D 그래프 설정
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# x, y축 데이터 설정 (pH 값)
x = red_density
y = white_density
z = np.zeros_like(x)  # 초기 z값은 0으로 설정

# z축에 p-value 값 추가 (모든 데이터 포인트에 동일한 p-value 사용)
z_t_test = np.full(len(x), p_value)
z_bartlett = np.full(len(y), bartlett_p_value)

# 3D 산포도 그리기
ax.scatter(x, y, z_t_test, color='r', label='T-Test p-value')
ax.scatter(x, y, z_bartlett, color='b', label='Bartlett Test p-value')

# 축 및 레이블 설정
ax.set_xlabel('Red Wine pH')
ax.set_ylabel('White Wine pH')
ax.set_zlabel('p-value')

ax.set_title('T-Test and Bartlett Test p-value Distribution')
ax.legend()

plt.show()
```
