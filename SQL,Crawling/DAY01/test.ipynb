```python











import random

# 윷을 던지는 함수
def throw_yut():
    sticks = [random.randint(0, 1) for _ in range(4)]  # 4개의 윷가락 랜덤으로 0 또는 1 생성
    count = sticks.count(1)  # 윷가락이 1인 개수를 센다
    score = 0
    if count == 4:
        score = 5  # 모 (5점)
    elif count == 0:
        score = 4  # 윷 (4점)
    elif count == 3:
        score = 3  # 걸 (3점)
    elif count == 2:
        score = 2  # 개 (2점)
    else:
        score = 1  # 도 (1점)
    return score

# 윷놀이 게임 함수
def yutnori_game():
    player1_score = 0
    player2_score = 0
    player_turn = 1  # 1이면 흥부 차례, 2이면 놀부 차례
    
    while player1_score < 20 and player2_score < 20:
        if player_turn == 1:
            print("흥부의 차례입니다.")
            score = throw_yut()
            player1_score += score
            print(f"흥부의 점수: {score} (총 {player1_score}점)")
            if score >= 4:
                if player1_score >= 20:
                    break  # 점수가 20점 이상이면 게임 종료
                continue  # 점수가 4점 이상이면 한 번 더 던진다
            player_turn = 2  # 다음 차례는 놀부
        else:
            print("놀부의 차례입니다.")
            score = throw_yut()
            player2_score += score
            print(f"놀부의 점수: {score} (총 {player2_score}점)")
            if score >= 4:
                if player2_score >= 20:
                    break  # 점수가 20점 이상이면 게임 종료
                continue  # 점수가 4점 이상이면 한 번 더 던진다
            player_turn = 1  # 다음 차례는 흥부

    if player1_score >= 20:
        print("흥부가 승리했습니다!")
    else:
        print("놀부가 승리했습니다!")

if __name__ == "__main__":
    yutnori_game()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import koreanize_matplotlib

def draw_two_plots(title, x_data, max_temp_list, min_temp_list):
    """
    주어진 데이터를 사용하여 두 개의 라인 그래프를 그립니다.
    
    Parameters:
    - title (str): 그래프의 제목
    - x_data (list): x축 데이터 (연도)
    - max_temp_list (list): 최고 기온 데이터
    - min_temp_list (list): 최저 기온 데이터
    """
    plt.figure(figsize=(15, 5))
    plt.plot(x_data, max_temp_list, marker='o', linestyle='-', color='red', label='최고 기온')
    plt.plot(x_data, min_temp_list, marker='s', linestyle='--', color='blue', label='최저 기온')
    
    plt.title(title)
    plt.legend()
    plt.show()

def main():
    # 시작 연도, 마지막 연도, 비교할 월을 입력받습니다.
    start_year = int(input("시작 연도를 입력하세요: "))
    end_year = int(input("마지막 연도를 입력하세요: "))
    search_month = int(input("기온 변화를 측정할 달을 입력하세요: "))
    
    # CSV 파일에서 데이터를 읽어옵니다.
    weather_df = pd.read_csv('daegu-utf8.csv', encoding='utf-8-sig')
    
    # 날짜 컬럼을 datetime 형식으로 변환합니다.
    weather_df['날짜'] = pd.to_datetime(weather_df['날짜'], format='%Y-%m-%d')
    
    # 시작 연도부터 마지막 연도까지의 연도 리스트를 생성합니다.
    years = list(range(start_year, end_year + 1))
    
    # 최고 기온과 최저 기온 평균을 저장할 리스트를 초기화합니다.
    max_temp_list = []
    min_temp_list = []
    
    # 각 연도별로 데이터를 필터링하여 최고 기온과 최저 기온 평균을 계산합니다.
    for year in years:
        monthly_data = weather_df[(weather_df['날짜'].dt.year == year) & (weather_df['날짜'].dt.month == search_month)]
        max_temp_list.append(round(monthly_data['최고기온(℃)'].mean(), 1))
        min_temp_list.append(round(monthly_data['최저기온(℃)'].mean(), 1))
    
    # 결과를 출력합니다.
    print(f'{start_year}년부터 {end_year}년까지 {search_month}월의 기온 변화')
    print(f'{search_month}월 최저기온 평균: {min_temp_list}')
    print(f'{search_month}월 최고기온 평균: {max_temp_list}')
    
    # 그래프를 그립니다.
    title = f'{start_year}년부터 {end_year}년까지 {search_month}월의 기온 변화'
    draw_two_plots(title, years, max_temp_list, min_temp_list)

# 프로그램을 실행합니다.
main()
```
