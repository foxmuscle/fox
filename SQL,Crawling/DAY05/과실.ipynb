```python

data=pd.read_excel('fruit.xlsx')
data

# 부산과 대구 데이터 추출-사과

ss = data[data['시도별'] == '서울특별시']
seoul = ss[['시점', '사과:생산량 (톤)']]
seoul['지역'] = '서울특별시'


b = data[data['시도별'] == '부산광역시']
busan = b[['시점', '사과:생산량 (톤)']]
busan['지역'] = '부산광역시'

d = data[data['시도별'] == '대구광역시']
daegu = d[['시점', '사과:생산량 (톤)']]
daegu['지역'] = '대구광역시'

i = data[data['시도별'] == '인천광역시']
incheon = i[['시점', '사과:생산량 (톤)']]
incheon['지역'] = '인천광역시'

g = data[data['시도별'] == '광주광역시']
gwangju = g[['시점', '사과:생산량 (톤)']]
gwangju['지역'] = '광주광역시'

d2 = data[data['시도별'] == '대전광역시']
daejeon = d2[['시점', '사과:생산량 (톤)']]
daejeon['지역'] = '대전광역시'

u = data[data['시도별'] == '울산광역시']
ulsan = u[['시점', '사과:생산량 (톤)']]
ulsan['지역'] = '울산광역시'

g2 = data[data['시도별'] == '경기도']
gyeonggi = g2[['시점', '사과:생산량 (톤)']]
gyeonggi['지역'] = '경기도'

g3 = data[data['시도별'] == '강원도']
gwd = g3[['시점', '사과:생산량 (톤)']]
gwd['지역'] = '강원도'

c = data[data['시도별'] == '충청북도']
chungcheongbuk = c[['시점', '사과:생산량 (톤)']]
chungcheongbuk['지역'] = '충청북도'

c2 = data[data['시도별'] == '충청남도']
chungcheongnam= c2[['시점', '사과:생산량 (톤)']]
chungcheongnam['지역'] = '충청남도'

j = data[data['시도별'] == '전라북도']
jeonbuk = j[['시점', '사과:생산량 (톤)']]
jeonbuk['지역'] = '전라북도'


j2 = data[data['시도별'] == '전라남도']
jeollanam = j2[['시점', '사과:생산량 (톤)']]
jeollanam['지역'] = '전라남도'

g4 = data[data['시도별'] == '경상북도']
gb = g4[['시점', '사과:생산량 (톤)']]
gb['지역'] = '경상북도'

g5 = data[data['시도별'] == '경상남도']
gyeongsangnam = g5[['시점', '사과:생산량 (톤)']]
gyeongsangnam['지역'] = '경상남도'

j3 = data[data['시도별'] == '제주도']
jeju = j3[['시점', '사과:생산량 (톤)']]
jeju['지역'] = '제주도'

```

```python
# 선그래프 만들기----경산북도 포함
df = pd.concat([seoul,busan, daegu,incheon,gwangju,daejeon,ulsan,gyeonggi,gwd,chungcheongbuk,chungcheongnam,
                jeonbuk,jeollanam,gb,gyeongsangnam,jeju])

def drawLineGraph(title, DF):
    colors = ['#1E90FF', '#FF8C00', '#FF7F50', '#FF6347', '#FF4500',
              '#FFD700', '#00BFFF', '#00CED1', '#00FA9A', '#32CD32',
              '#ADFF2F', '#9ACD32', '#DAA520', '#FF69B4', '#DDA0DD',
              '#8A2BE2', '#6A5ACD', '#9370DB', '#4682B4', '#B0E57C']

    area = DF['지역'].unique()
    plt.figure(figsize=(16, 9))

    for idx, a in enumerate(area):
        plt.plot(DF[DF['지역'] == a]['시점'], DF[DF['지역'] == a]['사과:생산량 (톤)'], label=a, color=colors[idx])
        # plt.plot(x, y, drawstyle='steps-post', label='steps-post')
        # plt.plot(x, y, 'o--', color='grey', alpha=0.3)

    plt.title(f'[{title}]', fontsize=20)
    plt.xlabel('연도' ,fontsize=15)
    plt.ylabel('사과:생산량 (톤)', fontsize=15)
    plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
    plt.xticks(x,rotation=45, fontsize=12)
    plt.yticks( fontsize=12)
    plt.tight_layout()
    plt.show()


drawLineGraph('전국 사과:생산량 (톤)', df)
```

```python
# 선그래프 만들기
df = pd.concat([seoul,busan, daegu,incheon,gwangju,daejeon,ulsan,gyeonggi,gwd,chungcheongnam,
                jeonbuk,jeollanam,gyeongsangnam,jeju])

def drawLineGraph(title, DF):
    colors = ['#1E90FF', '#FF8C00', '#FF7F50', '#FF6347', '#FF4500',
              '#FFD700', '#00BFFF', '#00CED1', '#00FA9A', '#32CD32',
              '#ADFF2F', '#9ACD32', '#DAA520', '#FF69B4', '#DDA0DD',
              '#8A2BE2', '#6A5ACD', '#9370DB', '#4682B4', '#B0E57C']

    area = DF['지역'].unique()
    plt.figure(figsize=(16, 9))

    for idx, a in enumerate(area):
        plt.plot(DF[DF['지역'] == a]['시점'], DF[DF['지역'] == a]['복숭아:생산량 (톤)'], label=a, color=colors[idx])
        # plt.plot(x, y, drawstyle='steps-post', label='steps-post')
        # plt.plot(x, y, 'o--', color='grey', alpha=0.3)

    plt.title(f'[{title}]', fontsize=20)
    plt.xlabel('연도' ,fontsize=15)
    plt.ylabel('복숭아:생산량 (톤)', fontsize=15)
    plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
    plt.xticks(x,rotation=45, fontsize=12)
    plt.yticks( fontsize=12)
    plt.tight_layout()
    plt.show()


drawLineGraph('전국 복숭아:생산량 (톤)', df)
```
