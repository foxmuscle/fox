```python
#  모듈 로딩 
import pandas as pd
```

```python
#  데이터 준비
file_name='수입차 통계(2014~2024).xlsx'
```

```python
#엑셀 파일 읽어와서 어떻게 생긴지 한번 보기  data  시트 부분 읽어보면 인덱스 부분이 Unnamed: 0 표기 된걸 볼수 있음
pd.read_excel(file_name, sheet_name='data')
 
```

```python
#변수명 =  시트 불러오기 =>sheet_name=None 을 쓰면 다불러오는거  ,위에필요 없는 3행 지우고 !)

df2013 = pd.read_excel(file_name, sheet_name='data', skiprows=3) .loc[:, ['Unnamed: 0', '2013']]#-->loc ,기준으로 
df2014 = pd.read_excel(file_name, sheet_name='data', skiprows=3).loc[:, ['Unnamed: 0', '2014']]
df2015 = pd.read_excel(file_name, sheet_name='Sheet1', skiprows=3).loc[:, ['Unnamed: 0', '2015']]
df2016 = pd.read_excel(file_name, sheet_name='Sheet3', skiprows=3).loc[:, ['Unnamed: 0', '2016']]
df2017 = pd.read_excel(file_name, sheet_name='Sheet4', skiprows=3).loc[:, ['Unnamed: 0', '2017']]
df2018 = pd.read_excel(file_name, sheet_name='Sheet5', skiprows=3).loc[:, ['Unnamed: 0', '2018']]
df2019 = pd.read_excel(file_name, sheet_name='Sheet6', skiprows=3).loc[:, ['Unnamed: 0', '2019']]
df2020 = pd.read_excel(file_name, sheet_name='Sheet7', skiprows=3).loc[:, ['Unnamed: 0', '2020']]
df2021 = pd.read_excel(file_name, sheet_name='Sheet8', skiprows=3).loc[:, ['Unnamed: 0', '2021']]
df2022 = pd.read_excel(file_name, sheet_name='Sheet9', skiprows=3).loc[:, ['Unnamed: 0', '2022']]
df2023 = pd.read_excel(file_name, sheet_name='Sheet10', skiprows=3).loc[:,['Unnamed: 0', '2023']]


#Unnamed 을 열을기준으로 브랜드로 바꾸기, 열을기준으로 
df2013.rename({'Unnamed: 0' : 'Brand'}, axis = 1, inplace = True) # axis = 0은 행, axis = 1은 열
df2014.rename({'Unnamed: 0' : 'Brand'}, axis = 1, inplace = True)
df2015.rename({'Unnamed: 0' : 'Brand'}, axis = 1, inplace = True)
df2016.rename({'Unnamed: 0' : 'Brand'}, axis = 1, inplace = True)
df2017.rename({'Unnamed: 0' : 'Brand'}, axis = 1, inplace = True)
df2018.rename({'Unnamed: 0' : 'Brand'}, axis = 1, inplace = True)
df2019.rename({'Unnamed: 0' : 'Brand'}, axis = 1, inplace = True)
df2020.rename({'Unnamed: 0' : 'Brand'}, axis = 1, inplace = True)
df2021.rename({'Unnamed: 0' : 'Brand'}, axis = 1, inplace = True)
df2022.rename({'Unnamed: 0' : 'Brand'}, axis = 1, inplace = True)
df2023.rename({'Unnamed: 0' : 'Brand'}, axis = 1, inplace = True)


# print(df2013)
# df2013
```

```python
#가져온 데이터 연결하기 concat
# axis=1열방향으로 병합하고 하고  기본은axis=0 => 행방향으로 병합

total_DF = pd.concat([df2013, df2014, df2015, df2016, df2017, df2018, 
                      df2019, df2020, df2021, df2022 ,df2023 ], axis = 1) 
#->  , join = 'inner'  -> 사용 병합 공통으로 존재하는 인덱스만 포함  -> 이거 쓰니깐 밑에 다 날라가서 기각                                                                                                                             

#잘 붙었나 확인
total_DF
```

```python
##데이터 프레임 ->병합 했으니깐 ! 열 이름 중복을 검사하기 . 열 이름이 앞서 나타났는지 아닌지 확인 Trun 아니면 False
#한번 확인해보기

print(total_DF.columns.duplicated())
```

```python
#병합 했으니깐 데이터프레임을 병합하거나 연결한 후 열 이름 중복을 검사하기 . 열 이름이 앞서 나타났는지 아닌지 확인 Trun 아니면 False
# ~ 뒤집기
total_DF = total_DF.loc[:, ~total_DF.columns.duplicated()]
total_DF
```
