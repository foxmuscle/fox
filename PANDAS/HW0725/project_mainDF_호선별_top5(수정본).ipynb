##### 연도별 역별 총합 하차인원 DF
- 1월 -12월 데이터 행 합쳐서 '총합' 칼럼 "년" 옆에 생성

```python
year_group_Df = rawDF.groupby("년")
year_group_Df.get_group(2006)
```

```python
sum_year = dict(zip( rawDF["년"].unique().tolist(), [0]*len(rawDF["년"].unique()) ) ) 

for year in rawDF["년"].unique():
    sum_year[year] = year_group_Df.get_group(year).sum()[2:].tolist()

# sum_year
```

```python
# 연도별 역별 하차인원 총합 DF 생성

year_out_DF = pd.DataFrame( sum_year, index=rawDF.columns[2:])
year_out_DF_final = year_out_DF.T.copy()
```

```python
year_out_DF_final.head()
```

```python
# 1호선, 2호선, 3호선 DF 구분
first_line = year_out_DF_final.loc[:,'설화명곡':'안심']
second_line = year_out_DF_final.loc[:, '문양':'영남대']
third_line = year_out_DF_final.loc[:, '칠곡경대병원':'용지']
```

##### 호선 별 2023 -2019 총 하차 이용객 TOP 5 그래프

```python
# 1호선 최근 5년간 하차승객 수 총합 TOP5 
# 막대 그래프
tmp = first_line.loc[2019:2023,:].sum(axis=0).sort_values(ascending=False)
tmp
```

```python
import my_font
my_font.set_costomfont()
```

```python

def station_tot(df_line):
    
    tmp = df_line.loc[2019:2023, :].sum(axis=0).sort_values(ascending=False)

    station = tmp[:5].index.tolist()
    total   = tmp[:5].tolist()

    return station, total


# x축, y축 데이터 생성
first_station, first_tot = station_tot(first_line)


second_station,second_tot  = station_tot(second_line)

third_station, third_tot  = station_tot(third_line)

color_list = ["red", "green", "gold"]

# 1호선
# plot
plt.figure( figsize=(8,6))

bar1 = plt.bar( first_station, first_tot, color="red")

# bar 상단 수치 
plt.bar_label(bar1, fmt ='{:,.0f}명', label_type='edge', padding=6, fontsize=12)

plt.title("최근 5년(2019-2023) 하차승객 TOP5 역사", fontsize=16, pad=15)
plt.ylim(bottom=0,top=max(first_tot)+ 10**7*0.5)

plt.ylabel("명", fontsize=12, labelpad=10)
plt.xlabel("역 이름", fontsize=12, labelpad=10)
plt.show()

# 2호선
# plot
plt.figure( figsize=(8,6))

bar2 = plt.bar( second_station, second_tot, color="green")

# bar 상단 수치 
plt.bar_label(bar2, fmt ='{:,.0f}명', label_type='edge', padding=6, fontsize=12)

plt.title("최근 5년(2019-2023) 하차승객 TOP5 역사", fontsize=16, pad=15)
plt.ylim(bottom=0,top=max(second_tot)+ 10**7*0.5)

plt.ylabel("명", fontsize=12, labelpad=10)
plt.xlabel("역 이름", fontsize=12, labelpad=10)
plt.show()

# 3호선
# plot
plt.figure( figsize=(8,6))

bar3 = plt.bar( third_station, third_tot, color="gold")

# bar 상단 수치 
plt.bar_label(bar3, fmt ='{:,.0f}명', label_type='edge', padding=6, fontsize=12)

plt.title("최근 5년(2019-2023) 하차승객 TOP5 역사", fontsize=16, pad=15)
plt.ylim(bottom=0,top=max(third_tot)+ 10**7*0.5)

plt.ylabel("명", fontsize=12, labelpad=10)
plt.xlabel("역 이름", fontsize=12, labelpad=10)
plt.show()

```
