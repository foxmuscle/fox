```python
#- 데이터 시각화 
# (1) 한글 처리
# 한글폰트 설정 => 폰트 매니저 모듈
from matplotlib import font_manager as fm 
from matplotlib import rc 

# 적용할 폰트 파일
FONT_FILE = r'C:\Windows\Fonts\NanumGothic.ttf'

# 폰트 패밀리 이름 가져오기
font_name=fm.FontProperties(fname=FONT_FILE).get_name()

# 새로운 폰트 패밀리 이름 지정
rc('font', family=font_name)
```

```python
fig=plt.figure(figsize=(12,10))
ax1, ax2=fig.subplots(1,2, sharey=True)

ax1.plot(moveTotalDF[:7])
import numpy as np
ax1.set_xticks(np.arange(0,7,1), moveTotalDF.index[:7].to_list(),rotation=45)
#ax1.set_xticklabels(moveTotalDF.index[:7].to_list(), rotation=45)

ax2.bar(moveTotalDF.index[:7].to_list(), moveTotalDF[:7])
ax2.set_xticks(np.arange(0,7,1), moveTotalDF.index[:7].to_list(),rotation=45)

fig.suptitle('[대구광역시 ==> 타광역시 이동인구수]', fontsize=20)
plt.show()
```

```python
fig=plt.figure(figsize=(12,10))
ax1, ax2=fig.subplots(1,2, sharey=True)

ax1.plot(moveTotalDF[7:])
import numpy as np
ax1.set_xticks(np.arange(0,9,1), moveTotalDF.index[7:].to_list(),rotation=45)
#ax1.set_xticklabels(moveTotalDF.index[7:].to_list(), rotation=45)

ax2.bar(moveTotalDF.index[7:].to_list(), moveTotalDF[7:])
ax2.set_xticks(np.arange(0,9,1), moveTotalDF.index[7:].to_list(),rotation=45)

fig.suptitle('[대구광역시 ==> 타도 이동인구수]', fontsize=20)
plt.show()
```

```python
# - 데이터 추출
moveTotalDF=cityDF.sum()
print(moveTotalDF)
```

```python
def drawGraphe():
    fig=plt.figure(figsize=(12,10))
    ax1, ax2=fig.subplots(2,1, sharex=True)

    ax1.plot(moveTotalDF)
    ax2.bar(moveTotalDF.index.to_list(),moveTotalDF)

    fig.suptitle('[대구광역시 ==> 타도 이동인구수]', fontsize=20)
    plt.show()
```

```python
drawGraphe()
```
